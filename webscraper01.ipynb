```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import string
```

```python
master_dict = {"contractor": [], "email": [], "website": []}
```

```python
uppercase_alphabet = string.ascii_uppercase
```

```python
for letter in uppercase_alphabet:
    url = f"https://www.gsaelibrary.gsa.gov/ElibMain/contractorList.do?contractorListFor={letter}"
    driver = webdriver.Chrome()
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    rows = soup.find_all("tbody")
    table_of_links = rows[-1]
    list_of_links = table_of_links.find_all("a")
    link_names = []
    for link in list_of_links:
        link_names.append(link.get_text(separator = " ", strip = True))
    print(link_names)
    print()
    print()
    for link_name in link_names[0:5]:
        try:
            contractor_link = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.LINK_TEXT, link_name)))
            contractor_link.click()
        except:
            link_name = link_name.split("(")[1]
            contractor_link = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT, link_name)))
            contractor_link.click()
        extract_contractor_information()
        driver.back()
    driver.quit()
        
```

```python
def extract_contractor_information():
    temp_list = []
    contractor_information_html = driver.page_source
    soup2 = BeautifulSoup(contractor_information_html, "html.parser")
    rows = soup2.find_all("tbody")
    contractor_information_table = rows[12]
    contractor_information = contractor_information_table.find_all("tr")
    for information in contractor_information:
        text = information.find_all("td")
        try:
            useful_info = text[1].text.strip()
            temp_list.append(useful_info)
        except:
            pass
    master_dict["contractor"].append(temp_list[1])
    master_dict["email"].append(temp_list[4])
    master_dict["website"].append(temp_list[5])
    
```

```python
contractor_link = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.LINK_TEXT, 'EOI, INC\n(DBA: MEDICAL RESOURCES)')))
contractor_link.click()
```
